### Model and method

```python
t = np.linspace(0, 4)
v1 = np.ones(t.shape)
v2 = -v1.copy()

v1p = (v1[0] + v2[0])/2

v1[t>2] = v1p
v2[t>2] = v1p

x1 = np.cumsum(v1*(t[1] - t[0]))
x2 = np.cumsum(v2*(t[1] - t[0]))
plt.plot(t, x1-2.1)
plt.plot(t, 2.01+x2)
plt.xlabel('time (s)')
plt.ylabel('x-position (m)')
```
